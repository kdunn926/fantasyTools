```python
"""
Forwards/Defencemen
-----------------------                            
Goals = 3pts
Assists = 2pts
Plus/Minus = 1pt
Penalty Minutes = 0.25pt
Powerplay Points = +0.5pt
Shots on Goal = 0.4pt

Goalies
-----------------------
Wins = 3pts
Goals Against = -1pt
Saves = 0.2pt
Shutouts = 2pts
""";

skaterScoringFields = ['goals','assists','plusMinus','penaltyMinutes','powerplayGoals','shots']
skaterScoring = pd.Series(index=skaterScoringFields, \
                          data=[3,2,1,0.25,0.5,0.4])

goalieScoringFields = ['wins', 'goalsAgainst', 'saves', 'shutouts']
goalieScoring = pd.Series(index=goalieScoringFields, \
                          data=[3,-1,0.2,2])

skaters['points'] = np.dot(skaters[skaterScoringFields], skaterScoring)
goalies['points'] = np.dot(goalies[goalieScoringFields], goalieScoring)
```

```python
goaliePerGameStats = pd.DataFrame(index=goalies.index.values)

def setGoaliePerGameStats():
    for stat in goalieScoringFields:
        goaliePerGameStats[stat] = goalies[stat]/goalies['gamesPlayed']
    goaliePerGameStats['pointsPerGame'] = goalies['points']/goalies['gamesPlayed']

setGoaliePerGameStats()

goalieZScores = pd.DataFrame(index=goalies.index.values)
goalieZScores['name'] = goalies['name']

def setGoalieZScores(goalie):
    for stat in goalieScoringFields:
        goalieZScores.loc[goalie, stat] = (goaliePerGameStats.loc[goalie, stat] - goaliePerGameStats[stat].mean())/goaliePerGameStats[stat].std(ddof=1)
    
    goalieZScores.loc[goalie, 'playerRank'] = np.mean([np.average(goalieZScores.loc[goalie, goalieScoringFields], \
                                                       weights=goalieScoring.tolist())])

goalieZScores.index.map(setGoalieZScores);
```

```python
from bs4 import BeautifulSoup

from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import WebDriverException

driver = webdriver.PhantomJS()

def ajaxComplete(driver):
    try:
        return 0 == driver.execute_script("return jQuery.active")
    except WebDriverException:
        pass

def getSalaries():
    
    positions = list()
    players = list()
    salaries = list()

    theUrl = 'https://www.fanduel.com/nextnhlgame' #?tableId=6637886'  

    driver.get(theUrl)

    #wait for ajax items to load
    WebDriverWait(driver, 10).until(ajaxComplete,  "Timeout waiting for page to load")

    html = driver.page_source

    soup = BeautifulSoup(html)

    table = soup.find('table', {'class': 'condensed player-list-table'}).find_all('tr') 
    #print table

    for row in table:
        cells = row.find_all("td")
        
        try:
            position = cells[0].get_text()

            player = cells[1].find('div').get_text()
            
            gamesPlayed = cells[3].get_text()
            
            if gamesPlayed == 0:
                continue

            salary = cells[5].get_text()[1:] #Skip the dollar sign

            positions.append(position.strip('\n\t'))
            players.append(player.strip('\n\t'))
            salaries.append(salary.strip(','))
            
        except:
            continue
    
    return pd.DataFrame(dict(position=positions, player=players, salary=salaries))

#contests = [(5862953,47516578), (5864129,47518354)]
results = getSalaries() #, ignore_index=True)
```

```python
skaterPerGameStats = pd.DataFrame(index=skaterSubset.index.values)

def setPerGameStats(player):
    for stat in skaterScoringFields:
        try:
            skaterPerGameStats.loc[player, stat] = float(skaters.loc[player, stat])/skaters.loc[player, 'gamesPlayed']
        except TypeError:
            print skaters.loc[player, 'name'] #, stat
            
skaterPerGameStats['name'] = skaterSubset['name']
skaterPerGameStats['position'] = skaterSubset['position']

skaterPerGameStats.index.map(setPerGameStats)

skaterPerGameStats['timeOnIce'] = skaterSubset['timeOnIce']
skaterPerGameStats['pointsPerGame'] = skaterSubset['points']/skaterSubset['gamesPlayed']
```

```python
skaterZScores = pd.DataFrame(index=skaterSubset.index.values)
skaterZScores['name'] = skaterSubset['name']
skaterZScores['position'] = skaterSubset['position']
skaterZScores['salary'] = skaterSubset['salary']


def setZScores(player):
    pos = skaterSubset.loc[player, 'position']
    for stat in skaterScoringFields:
        try:
            skaterZScores.loc[player, stat] = (skaterPerGameStats.loc[player, stat] - perPositionStats.loc[pos, (stat, 'mean')])/perPositionStats.loc[pos, (stat, 'std')]
        except:
            print skaterSubset.loc[player, 'name'], stat
            
    timeZScore = (skaterPerGameStats.loc[player, 'timeOnIce'] - perPositionStats.loc[pos, ('timeOnIce', 'mean')])/perPositionStats.loc[pos, ('timeOnIce', 'std')]
    skaterZScores.loc[player, 'timeOnIce'] = timeZScore
    
    pointsZScore = (skaterPerGameStats.loc[player, 'pointsPerGame'] - perPositionStats.loc[pos, ('pointsPerGame', 'mean')])/perPositionStats.loc[pos, ('pointsPerGame', 'std')]
    skaterZScores.loc[player, 'pointsPerGame'] = pointsZScore
    
    skaterZScores.loc[player, 'playerRank'] = np.mean([np.average(skaterZScores.loc[player , skaterScoringFields], \
                                                            weights=skaterScoring.tolist()), \
                                                 2*timeZScore, pointsZScore])

skaterZScores.index.map(setZScores);
```

```python
teamRanks = pd.DataFrame(index=unique(skaterSubset.team_id.values))

def computeTeamScores():
    for team in teamRanks.index.values:
        thisTeam = skaterSubset[skaters.team_id == team]
        teamRanks.loc[team, 'fScore'] =  -99
        for pos in set(unique(thisTeam.position.values)).difference(['D']): # Some teams rosters don't have both L/R wings
            thisPos = thisTeam[thisTeam.position == pos]
            bestPlayers = skaterZScores.loc[thisPos.index.values].sort(['playerRank'], ascending=False)
            if teamRanks.loc[team, 'fScore'] == -99:
                teamRanks.loc[team, 'fScore'] = np.mean([bestPlayers['playerRank'][0:3].values])
            else:
                teamRanks.loc[team, 'fScore'] = np.mean([teamRanks.loc[team, 'fScore'], np.mean([bestPlayers['playerRank'][0:3].values]) ])

            
        teamRanks.loc[team, 'dScore'] = -99
        thisPos = thisTeam[thisTeam.position == 'D']
        for dexterity in unique(thisPos.shoots.values): # Some teams may not have L+R defensive shooters
            tmp = thisPos[thisPos.shoots == dexterity]
            bestPlayers = skaterZScores.loc[tmp.index.values].sort(['playerRank'], ascending=False)
            if teamRanks.loc[team, 'dScore'] == -99: 
                teamRanks.loc[team, 'dScore'] = np.mean([bestPlayers['playerRank'][0:3].values])
            else:
                teamRanks.loc[team, 'dScore'] = np.mean([teamRanks.loc[team, 'dScore'], np.mean([bestPlayers['playerRank'][0:3].values]) ])

        tmp = goalies[goalies.team_id == team]
        bestGoalie = goalieZScores.loc[tmp.index.values].sort(['playerRank'], ascending=False)
        teamRanks.loc[team, 'gScore'] = bestGoalie['playerRank'].values[0]
            
computeTeamScores()
```

```python
#from bs4 import BeautifulSoup

#from selenium import webdriver
#from selenium.webdriver.support.ui import WebDriverWait
#from selenium.common.exceptions import WebDriverException

#driver = webdriver.PhantomJS()

def fanDuelToNhlId(team):
    if team == 'NJ':
        return 'NJD'
    elif team == 'MON':
        return 'MTL'
    elif team == 'SJ':
        return 'SJS'
    elif team == 'PHO':
        return 'ARI'
    elif team == 'TB':
        return 'TBL'
    elif team == 'WAS':
        return 'WSH'
    elif team == 'ANH':
        return 'ANA'
    elif team == 'CLS':
        return 'CBJ'
    elif team == 'LA':
        return 'LAK'
    else:
        return team

def ajaxComplete(driver):
    try:
        return 0 == driver.execute_script("return jQuery.active")
    except WebDriverException:
        pass

def getNextNHLGames():
    twoWay = dict()
    
    theUrl = 'https://www.fanduel.com/nextnhlgame'  

    driver.get(theUrl)

    # wait for ajax items to load
    WebDriverWait(driver, 10).until(ajaxComplete,  "Timeout waiting for page to load")

    html = driver.page_source

    soup = BeautifulSoup(html)

    gameList = soup.find('div', {'class': 'fixture-picker-button-container'}).find_all('span') 
    numGames = len(gameList)/3 # Three spans per game: away team, home team, start time

    for i in range(numGames):
        away = fanDuelToNhlId(gameList[i*3].get_text().strip(' \n\t').upper())
        home = fanDuelToNhlId(gameList[i*3+1].get_text().strip(' \n\t').upper())
        
        twoWay[away] = home
        twoWay[home] = away

    return twoWay

theGames = getNextNHLGames()

upcoming = theGames.keys()

relevantPlayers = skaters[skaters.team_id.isin(upcoming)][['position', 'team_id']]
relevantPlayers = relevantPlayers.append(goalies[goalies.team_id.isin(upcoming)][['pos', 'team_id']])
```
