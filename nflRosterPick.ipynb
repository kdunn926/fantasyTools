```python
# Turn the scoring fields into a column list
fieldList = scoringFields.split(',')

# Create a scoring vector/series for calculating points
pointSeries = pd.Series([0.04,4,-1,0.1,6,0.1,6,6,6,6,-2,0.5,2,2,1,1,1,2,2,2,6,6,2], index=fieldList, dtype=float)
pointSeries
```

```python
seasonMap = pd.read_sql("SELECT gsis_id, season_year, season_type, week, home_team from game", con)
seasonMap = seasonMap.set_index('gsis_id')

df = df.join(seasonMap)

career = df[df['season_type'] == 'Regular']

df = df[df['season_year'] == 2014]
df = df[df['season_type'] == 'Regular']

threeGames = df[df.week.isin(range(thisWeek-3, thisWeek))]
```

```python
df.drop('season_year', axis=1, inplace=True)
df.drop('season_type', axis=1, inplace=True)

df = df.reset_index()
career = career.reset_index()
threeGames = threeGames.reset_index()
```

```python
# Function to convert field goal yards to points
def yardsToPoints(yards):
    if yards <= 19:
        return 3
    elif yards <= 29:
        return 3
    elif yards <= 39:
        return 3
    elif yards <= 49:
        return 4
    elif yards > 49:
        return 5

# Apply the above function to replace field goal yards with points earned
df['kicking_fgm_yds'] = df['kicking_fgm_yds'].map(yardsToPoints)
career['kicking_fgm_yds'] = career['kicking_fgm_yds'].map(yardsToPoints)
threeGames['kicking_fgm_yds'] = threeGames['kicking_fgm_yds'].map(yardsToPoints)

df['score'] = np.dot(df[['passing_yds', 'passing_tds', 'passing_int', 'rushing_yds', 'rushing_tds', 'receiving_yds', 'receiving_tds', 'kickret_tds', 'puntret_tds', 'fumbles_rec_tds', 'fumbles_lost', 'receiving_rec', 'passing_twoptm', 'rushing_twoptm', 'kicking_xpmade', 'kicking_fgm_yds', 'defense_sk', 'defense_int', 'defense_frec', 'defense_safe', 'defense_frec_tds', 'defense_int_tds', 'defense_puntblk']], pointSeries)
career['score'] = np.dot(career[['passing_yds', 'passing_tds', 'passing_int', 'rushing_yds', 'rushing_tds', 'receiving_yds', 'receiving_tds', 'kickret_tds', 'puntret_tds', 'fumbles_rec_tds', 'fumbles_lost', 'receiving_rec', 'passing_twoptm', 'rushing_twoptm', 'kicking_xpmade', 'kicking_fgm_yds', 'defense_sk', 'defense_int', 'defense_frec', 'defense_safe', 'defense_frec_tds', 'defense_int_tds', 'defense_puntblk']], pointSeries)
threeGames['score'] = np.dot(threeGames[['passing_yds', 'passing_tds', 'passing_int', 'rushing_yds', 'rushing_tds', 'receiving_yds', 'receiving_tds', 'kickret_tds', 'puntret_tds', 'fumbles_rec_tds', 'fumbles_lost', 'receiving_rec', 'passing_twoptm', 'rushing_twoptm', 'kicking_xpmade', 'kicking_fgm_yds', 'defense_sk', 'defense_int', 'defense_frec', 'defense_safe', 'defense_frec_tds', 'defense_int_tds', 'defense_puntblk']], pointSeries)
```

```python
# Split the dataset into seperate home/away game sets
homeStats = df[df['team'] == df['home_team']][['gsis_id', 'player_id', 'score', 'receiving_tar']]
awayStats = df[df['team'] != df['home_team']][['gsis_id', 'player_id', 'score', 'receiving_tar']]

homeStatsCareer = career[career['team'] == career['home_team']][['gsis_id', 'player_id', 'score', 'receiving_tar']]
awayStatsCareer = career[career['team'] != career['home_team']][['gsis_id', 'player_id', 'score', 'receiving_tar']]

homeStatsRecent = threeGames[threeGames['team'] == threeGames['home_team']][['gsis_id', 'player_id', 'score', 'receiving_tar']]
awayStatsRecent = threeGames[threeGames['team'] != threeGames['home_team']][['gsis_id', 'player_id', 'score', 'receiving_tar']]
```

```python
# Aggregate points seperately for home/away games - season per-game, career per-game, recently per-game
def computePerGameStats(dataSet):
    tmp = dataSet.groupby(['player_id', 'gsis_id']).agg([np.sum, np.std])
    tmp = tmp[[('score', 'sum'),('receiving_tar', 'sum')]].reset_index()
    tmp.columns = ['player_id', 'gsis_id', 'score', 'targets']
    return tmp.groupby(['player_id']).agg([np.mean, np.std])

seasonPerGame = pd.DataFrame()
careerPerGame = pd.DataFrame()
recentPerGame = pd.DataFrame()
for theGroup in [['season', [homeStats, awayStats]], \
                 ['career', [homeStatsCareer, awayStatsCareer]], \
                 ['recent', [homeStatsRecent, awayStatsRecent]]]:
    dataSet = theGroup[0]
    [homeGames, awayGames] = theGroup[1]
    if dataSet == 'season':
        # Perform per game aggreation for home games and append to new DataFrame
        seasonPerGame = computePerGameStats(homeGames)
        # Perform per game aggreation for away games and append to new DataFrame
        seasonPerGame = seasonPerGame.join(computePerGameStats(awayGames), lsuffix='-home', rsuffix='-away')
    elif dataSet == 'career':
        # Perform per game aggreation for home games and append to new DataFrame
        careerPerGame = computePerGameStats(homeGames)
        # Perform per game aggreation for away games and append to new DataFrame
        careerPerGame = careerPerGame.join(computePerGameStats(awayGames), lsuffix='-home', rsuffix='-away')
    elif dataSet == 'recent':
        # Perform per game aggreation for home games and append to new DataFrame
        recentPerGame = computePerGameStats(homeGames)
        # Perform per game aggreation for away games and append to new DataFrame
        recentPerGame = recentPerGame.join(computePerGameStats(awayGames), lsuffix='-home', rsuffix='-away')
```

```python
# Aggregate accumulated points seperately for home/away games - {season, career, recently} cummulative and per-play average
playerGroupsHome = homeStats.groupby(['player_id']).agg([np.sum, np.std, 'count', np.mean])
playerGroupsAway = awayStats.groupby(['player_id']).agg([np.sum, np.std, 'count', np.mean])

playerGroupsHomeCareer = homeStatsCareer.groupby(['player_id']).agg([np.sum, np.std, 'count', np.mean])
playerGroupsAwayCareer = awayStatsCareer.groupby(['player_id']).agg([np.sum, np.std, 'count', np.mean])

playerGroupsHomeRecent = homeStatsRecent.groupby(['player_id']).agg([np.sum, np.std, 'count', np.mean])
playerGroupsAwayRecent = awayStatsRecent.groupby(['player_id']).agg([np.sum, np.std, 'count', np.mean])
```

```python
# Join the home/away sets into a compound set, applying suffixes to differentiate between them
playerGroups = playerGroupsHome.join(playerGroupsAway, lsuffix='-home', rsuffix='-away')

playerGroupsCareer = playerGroupsHomeCareer.join(playerGroupsAwayCareer, lsuffix='-home', rsuffix='-away')

playerGroupsRecent = playerGroupsHomeRecent.join(playerGroupsAwayRecent, lsuffix='-home', rsuffix='-away')
```

```python
# Filter for players with variance, i.e. drop records with few data points
playerGroups = playerGroups[playerGroups['score-home']['std'] > 0]
playerGroups = playerGroups[playerGroups['score-away']['std'] > 0]

playerGroupsCareer = playerGroupsCareer[playerGroupsCareer['score-home']['std'] > 0]
playerGroupsCareer = playerGroupsCareer[playerGroupsCareer['score-away']['std'] > 0]

playerGroupsRecent = playerGroupsRecent[playerGroupsRecent['score-home']['std'] > 0]
playerGroupsRecent = playerGroupsRecent[playerGroupsRecent['score-away']['std'] > 0]
```

```python
# Merge in the per-game stats and apply a suffix
playerGroups = playerGroups.join(seasonPerGame, rsuffix='-perGame')
playerGroupsCareer = playerGroupsCareer.join(careerPerGame, rsuffix='-perGame')
playerGroupsRecent = playerGroupsRecent.join(recentPerGame, rsuffix='-perGame')
```

```python
def setZScores(group, player, perPositionMeanAndStd):
    position = group.loc[player, 'position']
    
    # Compute seperate z-scores for player's home/away per-play performance
    group.loc[player, 'playHomeZ'] = (((group.loc[player][('score-home', 'mean')]) - (perPositionMeanAndStd.loc[position]['score-home']['mean']['mean'])) / (perPositionMeanAndStd.loc[position]['score-home']['mean']['std'])) - \
                                     ((group.loc[player][('score-home', 'std')] - perPositionMeanAndStd.loc[position]['score-home']['std']['mean']) / perPositionMeanAndStd.loc[position]['score-home']['std']['std']) 
    group.loc[player, 'playAwayZ'] = (((group.loc[player][('score-away', 'mean')]) - (perPositionMeanAndStd.loc[position]['score-away']['mean']['mean'])) / (perPositionMeanAndStd.loc[position]['score-away']['mean']['std'])) - \
                                     ((group.loc[player][('score-away', 'std')] - perPositionMeanAndStd.loc[position]['score-away']['std']['mean']) / perPositionMeanAndStd.loc[position]['score-away']['std']['std']) 

    # Compute seperate z-scores for player's home/away per-game performance
    group.loc[player, 'gameHomeZ'] = (((group.loc[player][('score-home-perGame', 'mean')]) - (perPositionMeanAndStd.loc[position]['score-home-perGame']['mean']['mean'])) / (perPositionMeanAndStd.loc[position]['score-home-perGame']['mean']['std'])) - \
                                     ((group.loc[player][('score-home-perGame', 'std')] - perPositionMeanAndStd.loc[position]['score-home-perGame']['std']['mean']) / perPositionMeanAndStd.loc[position]['score-home-perGame']['std']['std']) 
    group.loc[player, 'gameAwayZ'] = (((group.loc[player][('score-away-perGame', 'mean')]) - (perPositionMeanAndStd.loc[position]['score-away-perGame']['mean']['mean'])) / (perPositionMeanAndStd.loc[position]['score-away-perGame']['mean']['std'])) - \
                                     ((group.loc[player][('score-away-perGame', 'std')] - perPositionMeanAndStd.loc[position]['score-away-perGame']['std']['mean']) / perPositionMeanAndStd.loc[position]['score-away-perGame']['std']['std']) 

    # Compute seperate z-scores for player's home/away cummulative performance 
    group.loc[player, 'cummulativeHomeZ'] = (((group.loc[player][('score-home', 'sum')]) - (perPositionMeanAndStd.loc[position]['score-home']['sum']['mean'])) / (perPositionMeanAndStd.loc[position]['score-home']['sum']['std']))
    group.loc[player, 'cummulativeAwayZ'] = (((group.loc[player][('score-away', 'sum')]) - (perPositionMeanAndStd.loc[position]['score-away']['sum']['mean'])) / (perPositionMeanAndStd.loc[position]['score-away']['sum']['std']))

    # Could add average per-play targets by using total offensive plays in each game and receivers targets for that game
    #
    #
    
    # Compute seperate z-scores for player's per-game targets
    group.loc[player, 'targetsHomeZ'] = (((group.loc[player][('targets-home', 'mean')]) - (perPositionMeanAndStd.loc[position]['targets-home']['mean']['mean'])) / (perPositionMeanAndStd.loc[position]['targets-home']['mean']['std'])) - \
                                     ((group.loc[player][('targets-home', 'std')] - perPositionMeanAndStd.loc[position]['targets-home']['std']['mean']) / perPositionMeanAndStd.loc[position]['targets-home']['std']['std']) 
    group.loc[player, 'targetsAwayZ'] = (((group.loc[player][('targets-away', 'mean')]) - (perPositionMeanAndStd.loc[position]['targets-away']['mean']['mean'])) / (perPositionMeanAndStd.loc[position]['targets-away']['mean']['std'])) - \
                                     ((group.loc[player][('targets-away', 'std')] - perPositionMeanAndStd.loc[position]['targets-away']['std']['mean']) / perPositionMeanAndStd.loc[position]['targets-away']['std']['std']) 

    # Compute seperate z-scores for player's cummulative targets
    group.loc[player, 'totalTargetsHomeZ'] = (((group.loc[player][('receiving_tar-home', 'sum')]) - (perPositionMeanAndStd.loc[position]['receiving_tar-home']['sum']['mean'])) / (perPositionMeanAndStd.loc[position]['receiving_tar-home']['sum']['std']))
    group.loc[player, 'totalTargetsAwayZ'] = (((group.loc[player][('receiving_tar-away', 'sum')]) - (perPositionMeanAndStd.loc[position]['receiving_tar-away']['sum']['mean'])) / (perPositionMeanAndStd.loc[position]['receiving_tar-away']['sum']['std']))


for group in [playerGroups, playerGroupsCareer, playerGroupsRecent]:
    positionScores = group.groupby('position').agg([np.mean, np.std])
    for player in group.index.values:
        setZScores(group, player, positionScores)
```

```python
# Create a dataframe mapping between team names and id's
teamToInt = pd.DataFrame({'Team': df.team.unique(), 'TeamID':range(len(df.team.unique()))})

# Define some helper functions to easily query the dataframe map
def getTeamId(name):
    try:
        return int(teamToInt[teamToInt.Team == str(name)]['TeamID'])
    except:
        print name
    
intToTeam = teamToInt.set_index('TeamID')

def getTeamName(id):
    return str(intToTeam.loc[int(id)]['Team'])
```

```python
# Load in the 2014 schedule
schedule = pd.read_csv('2014schedules.csv')

# Parse the extracted schedule into team ID's instead of names
def opponentNameToID(team):
    # Edge case for differences in naming
    if team == 'WSH':
        team = 'WAS'
    elif team == '@WSH':
        team = '@WAS'
    elif team == 'JAX':
        team = 'JAC'
    elif team == '@JAX':
        team = '@JAC'
    elif team == 'BYE':
        return np.NAN
    
    # Set team ID to negative if it's an away game
    if team[0] == '@':
        return -1*getTeamId(team[1:])
    else:
        return getTeamId(team)

# Apply the parsing routine week by week
for i in range(1, 18):
    schedule[str(i)] = schedule[str(i)].map(opponentNameToID)

# Fix a couple label inconsistencies
schedule['TEAM'][schedule.TEAM == 'JAX'] = 'JAC'
schedule['TEAM'][schedule.TEAM == 'WSH'] = 'WAS'

schedule = schedule.set_index('TEAM')

# Primitive function for indexing into the schedule table
def getOpponentId(team, week):
    return schedule.loc[str(team)][str(week)]
```

```python
# Create a game map in order to later associate stats with the opposing defenses
gameMaps = pd.read_sql("SELECT gsis_id, home_team, away_team from game", con)
gameMaps = gameMaps.merge(teamToInt, left_on='home_team', right_on='Team')
gameMaps = gameMaps.merge(teamToInt, left_on='away_team', right_on='Team')
gameMaps.drop(['Team_x', 'Team_y'], axis=1, inplace=True)
gameMaps = gameMaps.set_index('gsis_id')
```

```python
# Set the scoring fields for the SQL query
scoringFields = "passing_yds, passing_tds, passing_int, rushing_yds, rushing_tds, receiving_yds, receiving_tds, kickret_tds, puntret_tds, fumbles_rec_tds, fumbles_lost, receiving_rec, passing_twoptm, rushing_twoptm, kicking_xpmade, kicking_fgm_yds, defense_sk, defense_int, defense_frec, defense_safe, defense_frec_tds, defense_int_tds, defense_puntblk"

allFields3 = "gsis_id, team, player_id, " + scoringFields
detail = pd.read_sql("SELECT " + allFields3 + " from play_player", con)

# Create a season map to filter only games of interest (e.g. Regular Season, 2013)
seasonMap = pd.read_sql("SELECT gsis_id, season_year, season_type, week, home_team from game", con)
seasonMap = seasonMap.set_index('gsis_id')

# Append the additional game data to the detail
detail = detail.set_index("gsis_id")
detail = detail.join(seasonMap)

# Apply the game-type/year filtering
detail = detail[detail['season_year'] == 2014]
detail = detail[detail['season_type'] == 'Regular']

# Drop some implicit data
detail.drop('season_year', axis=1, inplace=True)
detail.drop('season_type', axis=1, inplace=True)

# Reset the index for later aggregation
detail = detail.reset_index()

# Append the team ID data to the detail
detail = detail.merge(teamToInt, left_on='team', how='left', right_on='Team')

# Merge in positions for later groupby aggregation
detail = detail.merge(playerDetails[['position']], left_on='player_id', how='left', right_index=True)

# Calculate the per-game, by team stats
dSum = detail[detail.position.isin(['QB', 'RB', 'WR', 'TE'])].groupby(['gsis_id', 'TeamID', 'position']).agg(np.sum) 

# Calculate the per-play, by team stats
dAvg = detail[detail.position.isin(['QB', 'RB', 'WR', 'TE'])].groupby(['gsis_id', 'TeamID', 'position']).agg([np.mean, 'count']) 
```

```python
dAccrued = dSum.copy()

# Convert perspective : offensive stats accrued become defensive stats allowed 
def setOpponent(game):
    if gameMaps.loc[game[0]]['TeamID_x'] == game[1]:
        return gameMaps.loc[game[0]]['TeamID_y']
    else:
        return gameMaps.loc[game[0]]['TeamID_x']

# Apply the map to associate stats with the opposing defense
dAccrued['oppTeamID'] = dAccrued.index.map(setOpponent)

# Drop the index to allow merging home_team data
dAccrued = dAccrued.reset_index() #index.droplevel('gsis_id')

# Append home team ID to the set
dAccrued = dAccrued.merge(seasonMap[['home_team']], left_on='gsis_id', right_index=True, how='left')

dAccrued.home_team = dAccrued.home_team.map(getTeamId)

# Allow the groupby to use position as second field
dAccrued = dAccrued.reset_index()

# Apply the final aggregation, yielding stats allowed from a defensive perspective, seperated by home/away games 
dAccruedHome = dAccrued[dAccrued.oppTeamID == dAccrued.home_team].groupby(['oppTeamID', 'position']).agg([np.mean, np.std])
dAccruedAway = dAccrued[dAccrued.oppTeamID != dAccrued.home_team].groupby(['oppTeamID', 'position']).agg([np.mean, np.std])

dAccruedHome = dAccruedHome.drop(['index', 'TeamID', 'week', 'home_team'], axis=1)
dAccruedAway = dAccruedAway.drop(['index', 'TeamID', 'week', 'home_team'], axis=1)
```

```python
def appendMean(f): return (f, str('mean'))
def appendStd(f): return (f, str('std'))

pointFields = ['passing_yds', 'passing_tds', 'passing_int', \
               'rushing_yds', 'rushing_tds', \
               'receiving_yds', 'receiving_tds', \
               'kickret_tds', 'puntret_tds', \
               'fumbles_rec_tds', 'fumbles_lost', \
               'receiving_rec', \
               'passing_twoptm', 'rushing_twoptm', \
               'kicking_xpmade', 'kicking_fgm_yds', \
               'defense_sk', 'defense_int', 'defense_frec', 'defense_safe', \
               'defense_frec_tds', 'defense_int_tds', 'defense_puntblk']
pointFields2 = pointFields

pointFields = map(appendMean, pointFields)
pointFields2 = map(appendStd, pointFields)

for thisSet in [dAccruedHome, dAccruedAway]:
    def setRowMean(defenseAndPosition):
        return np.dot(thisSet.loc[defenseAndPosition][pointFields], pointSeries)

    thisSet['pointMean'] = thisSet.index.map(setRowMean)

    def setRowStd(defenseAndPosition):
        return np.dot(thisSet.loc[defenseAndPosition][pointFields2], pointSeries)

    thisSet['pointStd'] = thisSet.index.map(setRowStd)
```

```python
# Join the home/away sets into a compound set, applying suffixes to differentiate between them
dAccrued = dAccruedHome.join(dAccruedAway, lsuffix='-home', rsuffix='-away')
```

```python
dPositionFinal = dAccrued[['pointMean-home', 'pointMean-away']]
dPositionFinal.columns = ['pointMean-home', 'pointMean-away']
def setOverall(index):
    return np.mean([dPositionFinal.loc[index, 'pointMean-home'], dPositionFinal.loc[index, 'pointMean-away']])

dPositionFinal['pointMean-overall'] = dPositionFinal.index.map(setOverall)
```

```python
def setPositionZScores(group, teamAndPos, perPositionMeanAndStd):
    position = teamAndPos[1] #group.loc[team, 'position']
    
    # Compute seperate z-scores for player's home/away per-play performance
    group.loc[teamAndPos, 'homeZ'] = (((group.loc[teamAndPos, 'pointMean-home']) - (perPositionMeanAndStd.loc[position]['pointMean-home']['mean'])) / (perPositionMeanAndStd.loc[position]['pointMean-home']['std']))
    group.loc[teamAndPos, 'awayZ'] = (((group.loc[teamAndPos, 'pointMean-away']) - (perPositionMeanAndStd.loc[position]['pointMean-away']['mean'])) / (perPositionMeanAndStd.loc[position]['pointMean-away']['std']))
    group.loc[teamAndPos, 'overallZ'] = (((group.loc[teamAndPos, 'pointMean-overall']) - (perPositionMeanAndStd.loc[position]['pointMean-overall']['mean'])) / (perPositionMeanAndStd.loc[position]['pointMean-overall']['std']))

    
for group in [dPositionFinal]: 
    positionScores = group.reset_index().groupby(['position']).agg([np.mean, np.std])
    
    for teamAndPos in group.index.values:
        setPositionZScores(group, teamAndPos, positionScores)
```

```python
# This scapes football outsiders for player defense-adjusted rankings based on their proprietary methods

from urllib2 import urlopen
from bs4 import BeautifulSoup   

def getIdFor(player, position, team):
    if player == 'Jo.Brown':
        return '00-0031051'
    elif player == 'Ju.Brown':
        return '00-0030466'
    elif player == 'R.Mathews':
        return '00-0027864'
    elif player == 'EJ Manuel':
        return '00-0030526'
    else:
        tempMap = nameToIdMap[nameToIdMap.position == position]
        tempMap = tempMap[tempMap.team == team]
        return tempMap.loc[player, 'player_id']
    
def getStatsFor(position, cells):
    
    player = cells[0].get_text()
    team = cells[1].get_text()
    
    dyar = cells[2].get_text()
    dyarRank = cells[3].get_text()
    dvoa = cells[6].get_text()
    dvoaRank = cells[7].get_text()
    voa = cells[8].get_text()
    
    if position == 'QB':
        qbRating = cells[9].get_text()

    return dict(playerId=getIdFor(player, position, team),
                name=player,
                position=position,
                team=team,
                dyar=np.float(dyar),
                dyarRank=np.int(dyarRank),
                dvoa=np.float(dvoa[0:-1]),
                dvoaRank=np.int(dvoaRank),
                voa=np.float(voa[0:-1])
                )
    

def getRanksFor(position):
    rankData = urlopen('http://www.footballoutsiders.com/stats/' + str(position.lower()))
    rankString = rankData.read()
    rankData.close()

    soup = BeautifulSoup(rankString)
    table = soup.find('table', {'class': 'stats'}).find_all('tr')

    thisPosition = pd.DataFrame()
    for row in table:
        cells = row.find_all("td")
        player = cells[0].get_text()
        # Skip the header row, it is implicit in the dataframe
        if player == 'Player':
            continue
        team = cells[1].get_text()
        
        if team == '2TM':
            continue
        if player == 'J.Jernigan' or \
           player == 'M.Lewis' or \
           player == 'J.Dwyer' or \
           player == 'A.Peterson' or \
           player == 'M.Cassel' or \
           player == 'D.Pitta' :
            continue

        #try:
        thisPosition = thisPosition.append(getStatsFor(position, cells), ignore_index=True)
        #except:
        #    print "Failed setting for:", player, "position:", position

    thisPosition = thisPosition.set_index('playerId')
    return thisPosition

dvoaDvarRanks = pd.DataFrame()

for position in ['QB', 'RB', 'WR', 'TE']:
    thisPosition = getRanksFor(position)
    dvoaDvarRanks = pd.concat([dvoaDvarRanks, thisPosition])
```

```python
playerGroups = playerGroups.set_index('player_id')
playerGroupsCareer = playerGroupsCareer.set_index('player_id')
playerGroupsRecent = playerGroupsRecent.set_index('player_id')

perPositionMeanAndStd2 = dvoaDvarRanks.groupby('position').agg([np.mean, np.std])

def setZScores2(player):
    try:
        position = dvoaDvarRanks.loc[player, 'position']

        # Compute seperate z-scores for player's home/away per-play performance
        dvoaDvarRanks.loc[player, 'dyarZ'] = ((dvoaDvarRanks.loc[player]['dyar']) - (perPositionMeanAndStd2.loc[position]['dyar']['mean'])) / (perPositionMeanAndStd2.loc[position]['dyar']['std'])

        # Compute seperate z-scores for player's home/away overall performance 
        dvoaDvarRanks.loc[player, 'dvoaZ'] = ((dvoaDvarRanks.loc[player]['dvoa']) - (perPositionMeanAndStd2.loc[position]['dvoa']['mean'])) / (perPositionMeanAndStd2.loc[position]['dvoa']['std'])
    except:
        dvoaDvarRanks.loc[player, 'dyarZ'] = np.nan
        dvoaDvarRanks.loc[player, 'dvoaZ'] = np.nan
        

dvoaDvarRanks.index.map(setZScores2);
```

```python
def getDefRanks():
    rankData = urlopen('http://www.footballoutsiders.com/stats/teamdef')
    
    rankString = rankData.read()
    rankData.close()

    soup = BeautifulSoup(rankString)
    table = soup.find('table', {'class': 'stats'}).find_all('tr')

    defRanks = pd.DataFrame()
    for row in table:
        cells = row.find_all("td")
        defTeam = cells[1].get_text()
        defDvoa = cells[2].get_text()
        
        # Skip the header rows, they are implicit in the dataframe
        if defTeam == 'Team' or len(cells[0].get_text()) == 0 :
            continue
            
        defPass = cells[6].get_text()
        defRush = cells[8].get_text()
        
        defRanks = defRanks.append(dict(teamId=getTeamId(defTeam),
                                        dvoa=np.float(defDvoa[0:-1]),
                                        passOppose=np.float(defPass[0:-1]),
                                        runOppose=np.float(defRush[0:-1])
                                        ), ignore_index=True)
        

    defRanks = defRanks.set_index('teamId')
    return defRanks
    
    
defenseDvoaRanks = getDefRanks()
```

```python
defenseTeamZs = pd.DataFrame()

defenseTeamZs['overallZ'] = ( defenseDvoaRanks['dvoa'] - defenseDvoaRanks['dvoa'].mean() ) / defenseDvoaRanks['dvoa'].std(ddof=1)
defenseTeamZs['passRushZ'] = ( defenseDvoaRanks['passOppose'] - defenseDvoaRanks['passOppose'].mean() ) / defenseDvoaRanks['passOppose'].std(ddof=1)
defenseTeamZs['runStopZ'] = ( defenseDvoaRanks['runOppose'] - defenseDvoaRanks['runOppose'].mean() ) / defenseDvoaRanks['runOppose'].std(ddof=1)
```

```python
def getLineRanksFor(line):
    rankData = urlopen('http://www.footballoutsiders.com/stats/' + str(line.lower()))
    rankString = rankData.read()
    rankData.close()

    soup = BeautifulSoup(rankString)
    table = soup.find('table', {'class': 'stats'}).find_all('tr')

    runBlock = pd.DataFrame()
    passProtect = pd.DataFrame()
    for row in table:
        cells = row.find_all("td")
        runBlockRank = cells[0].get_text()
        runBlockTeam = cells[1].get_text()
        # Skip the header rows, they are implicit in the dataframe
        if runBlockRank == 'RUN BLOCKING' or len(runBlockRank) == 0 or runBlockRank == 'x':
            continue
        aly = cells[2].get_text()
        rbYards = cells[3].get_text()
        powerSuccess = cells[4].get_text()
        powerRank = cells[5].get_text()
        stuffed = cells[6].get_text()
        stuffedRank = cells[7].get_text()
        level2Yards = cells[8].get_text()
        level2Rank = cells[9].get_text()
        openFieldYards = cells[10].get_text()
        openFieldRank = cells[11].get_text()

        runBlock = runBlock.append(dict(teamId=getTeamId(runBlockTeam),
                                        adjustedLineYards=np.float(aly),
                                        rbYards=np.float(rbYards),
                                        powerSuccess=np.float(powerSuccess[0:-1]),
                                        powerRank=np.int(powerRank),
                                        stuffed=np.int(stuffed[0:-1]),
                                        stuffedRank=np.int(stuffedRank),
                                        level2Yards=np.float(level2Yards),
                                        level2Rank=np.int(level2Rank),
                                        openFieldYards=np.float(openFieldYards),
                                        openFieldRank=np.int(openFieldRank)
                                        ), ignore_index=True)
        
        passProtectTeam = cells[12].get_text()
        passProtectRank = cells[13].get_text()
        sacks = cells[14].get_text()
        adjustedSackRate = cells[15].get_text()

        passProtect = passProtect.append(dict(teamId=getTeamId(passProtectTeam),
                                        rank=np.int(passProtectRank),
                                        sacks=np.int(sacks),
                                        adjustedSackRate=np.float(adjustedSackRate[0:-1])
                                        ), ignore_index=True)
        

    runBlock = runBlock.set_index('teamId')
    passProtect = passProtect.set_index('teamId')
    return runBlock, passProtect

adjustedLineYardsRanks = pd.DataFrame()

for line in ['ol', 'dl']:
    runBlock, passProtect = getLineRanksFor(line)
    composite = runBlock.join(passProtect)
    if line == 'dl':
        adjustedLineYardsRanks = adjustedLineYardsRanks.join(composite, lsuffix='-Offense', rsuffix='-Defense')
    else:
        adjustedLineYardsRanks = pd.concat([adjustedLineYardsRanks, composite])
```

```python
lineZs = pd.DataFrame()

lineZs['passProtectZ'] = (-adjustedLineYardsRanks['adjustedSackRate-Offense'] + adjustedLineYardsRanks['adjustedSackRate-Offense'].mean() ) / adjustedLineYardsRanks['adjustedSackRate-Offense'].std(ddof=1)
lineZs['passRushZ'] = (-adjustedLineYardsRanks['adjustedSackRate-Defense'] + adjustedLineYardsRanks['adjustedSackRate-Defense'].mean() ) / adjustedLineYardsRanks['adjustedSackRate-Defense'].std(ddof=1)

lineZs['runBlockZ'] = (adjustedLineYardsRanks['adjustedLineYards-Offense'] - adjustedLineYardsRanks['adjustedLineYards-Offense'].mean() ) / adjustedLineYardsRanks['adjustedLineYards-Offense'].std(ddof=1)
lineZs['runStopZ'] = (adjustedLineYardsRanks['adjustedLineYards-Defense'] - adjustedLineYardsRanks['adjustedLineYards-Defense'].mean() ) / adjustedLineYardsRanks['adjustedLineYards-Defense'].std(ddof=1)
```

```python
clusterData = pd.DataFrame()

def getDefenseMetrics2(position):
    if position == 'QB':
        return "passRushZ"
    elif position == 'RB':
        return "runStopZ"
    elif position == 'WR':
        #return 'targetsZ'
        return "passRushZ"
    elif position == 'TE':
        #return 'targetsZ'
        return "passRushZ"
    elif position == 'K':
        #return "specialZ"
        return "sumZ"
    else:
        print "Unknown position for scoring"
        
def getDefenseMetricsNew(teamId, position, homeOrAway):
    location = '{}Z'.format(homeOrAway)
    cummulative = 'overallZ' #.format(homeOrAway)
        
    metrics = list()
    # Append the position score conditioned on home/away
    metrics.append(dPositionFinal.loc[(teamId, position)][location])
    
    # Append the overall position score
    metrics.append(dPositionFinal.loc[(teamId, position)][cummulative])

    # Append the defense score conditioned on home/away
    metrics.append(dOverallFinal.loc[teamId][location])
    
    # Append the overall defense score
    metrics.append(dOverallFinal.loc[teamId][cummulative])
    
    return metrics
        
def getTeamLineMetric(position):
    if position == 'RB':
        return "runBlockZ"
    elif position == 'QB':
        return "passProtectZ"
    elif position == 'WR':
        return "passProtectZ"
    elif position == 'TE':
        return "passProtectZ"
    
def getTargetScore(player, homeOrAway):    
    perGameString = 'targets{}Z'.format(homeOrAway)
    cummulativeString = 'totalTargets{}Z'.format(homeOrAway)
    
    try:
        # Last 3 games
        recentScore = np.nanmean([playerGroupsRecent.loc[player][perGameString],\
                               playerGroupsRecent.loc[player][cummulativeString]])
    
    except:
        recentScore = np.nan
        
    # This season
    score = np.nanmean([playerGroups.loc[player][perGameString],\
                     playerGroups.loc[player][cummulativeString],\
    # Total career
                     playerGroupsCareer.loc[player][perGameString],\
                     playerGroupsCareer.loc[player][cummulativeString]])
    
    if np.isnan(recentScore):
        return score
    else:
        return np.nanmean([score, recentScore])

def getPlayerBlendedScore(player, position, teamId, homeOrAway):
    playerBlendedScore = 0
    
    targetScore = 0
    if position in ['RB', 'WR', 'TE']:
        targetScore = getTargetScore(player, homeOrAway)
    
    perPlay = 'play{}Z'.format(homeOrAway)
    perGame = 'game{}Z'.format(homeOrAway)
    cummulative = 'cummulative{}Z'.format(homeOrAway)

    try:
        recentScore = np.nanmean([playerGroupsRecent.loc[player][perPlay]/2, \
                               2*playerGroupsRecent.loc[player][perGame], \
                               playerGroupsRecent.loc[player][cummulative]])
        
    except:
        recentScore = np.nan
    
    try:
        # Compute average ranking of per-play, overall, and offensive line ranks 
        playerBlendedScore = np.nanmean([targetScore,\
                              dvoaDvarRanks.loc[player]['dvoaZ'],\
                              dvoaDvarRanks.loc[player]['dyarZ'],\
                              playerGroups.loc[player][perPlay]/2,\
                              2*playerGroups.loc[player][perGame],\
                              playerGroups.loc[player][cummulative],\
                              playerGroupsCareer.loc[player][perPlay],\
                              playerGroupsCareer.loc[player][perGame],\
                              playerGroupsCareer.loc[player][cummulative],\
                              2*lineZs.loc[teamId, getTeamLineMetric(position)] ])

        clusterData.loc[player, 'oPerPlay'] = playerGroups.loc[player][perPlay]
        clusterData.loc[player, 'oPerGame'] = playerGroups.loc[player][perGame]
        clusterData.loc[player, 'oCummulative'] = playerGroups.loc[player][cummulative]
        clusterData.loc[player, 'oLine'] = lineZs.loc[teamId, getTeamLineMetric(position)]
        clusterData.loc[player, 'dvoa'] = dvoaDvarRanks.loc[player]['dyoaZ']
        clusterData.loc[player, 'dvar'] = dvoaDvarRanks.loc[player]['dyarZ']

    except:
        # Compute average ranking of per-play or targets ranks and per-game rank
        playerBlendedScore = np.nanmean([targetScore,\
                              playerGroups.loc[player][perPlay]/2,\
                              2*playerGroups.loc[player][perGame],\
                              playerGroups.loc[player][cummulative],\
                              playerGroupsCareer.loc[player][perPlay],\
                              playerGroupsCareer.loc[player][perGame],\
                              playerGroupsCareer.loc[player][cummulative],\
                              2*lineZs.loc[teamId, getTeamLineMetric(position)] ]) 
        
        clusterData.loc[player, 'oPerPlay'] = playerGroups.loc[player][perPlay]
        clusterData.loc[player, 'oPerGame'] = playerGroups.loc[player][perGame]
        clusterData.loc[player, 'oCummulative'] = playerGroups.loc[player][cummulative]
        clusterData.loc[player, 'oLine'] = lineZs.loc[teamId, getTeamLineMetric(position)]      
        
    if np.isnan(recentScore):
        
        return playerBlendedScore
    else:
        
        return np.nanmean([playerBlendedScore, recentScore])

def getDefenseBlendedScore(opponentId, position, player):
    homeOrAway = 'away'
    # If the player is away (negative opponent), the defense is home
    if opponentId < 0:
            homeOrAway = 'home'
            opponentId = np.abs(opponentId)
            
    opponentId = int(opponentId)
    
    # The matchup determination is scored using only position-relevant defensive metrics
    numericMetrics = getDefenseMetricsNew(opponentId, position, homeOrAway)
    
    # The matchup determination is scored using only position-relevant defensive metrics
    fields2 = getDefenseMetrics2(position)

    clusterData.loc[player, 'dPoints'] = np.nanmean(numericMetrics)
    clusterData.loc[player, 'dLine'] = lineZs.loc[opponentId, fields2]
    clusterData.loc[player, 'dPosition'] = defenseTeamZs.loc[opponentId, fields2]
    clusterData.loc[player, 'dOverall'] = defenseTeamZs.loc[opponentId, 'overallZ']

    defenseBlendedScore = np.nanmean([np.nanmean(numericMetrics),\
                           lineZs.loc[opponentId, fields2],\
                           defenseTeamZs.loc[opponentId, fields2],\
                           2*defenseTeamZs.loc[opponentId, 'overallZ'] ])
    
    return defenseBlendedScore
    
week = thisWeek
def setMatchupScore2(player):
    # Lookup this player's team
    team = playerGroups.loc[player]['team']
    
    # Lookup this player's position
    position = playerGroups.loc[player]['position']
    
    # Lookup this player's team ID
    teamId = getTeamId(team)
    
    # Find this player's opposing team
    opponentId = getOpponentId(team, week)
    
    # A NaN implies this player has a bye week
    if np.isnan(opponentId):
        return (np.nan, np.nan)
    else:
        homeOrAway = 'Home'
        if opponentId < 0:
            homeOrAway = 'Away'
        
        playerBlendedScore = getPlayerBlendedScore(player, position, teamId, homeOrAway)

        defenseBlendedScore = getDefenseBlendedScore(opponentId, position, player)

        return (defenseBlendedScore, playerBlendedScore)
        
playerGroups['adjustedMatchupZs'] = playerGroups.index.map(setMatchupScore2)

#setMatchupScore2('00-0027942')

# Avoid nan tuples throwing off the alignment for plotting -- took a while to track down
playerGroups = playerGroups[~playerGroups.adjustedMatchupZs.isin([(np.nan, np.nan)])]
```
